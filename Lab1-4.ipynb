````python
import pandas as pd
````

````python
df = pd.read_csv('data/russian-housing-market.csv')# Load dataset here
````

1. Посчитайте количество записей в датасете с отсутствующей информацией о жилой площади жилого помещения

````python
no_life_sq = df['life_sq'].isna().sum()
print(no_life_sq)
````

2. Предложите способ заменить отсутствующие значения из пункта 1 и реализуйте его. Способ нужно аргументировать(сказать почему это достаточно хорошее приближение)

````python
# возьмем все записи с известными жилыми и общими площадями. Посчитаем общую пропорцию жилой площади к общей. Для всех отсутствующий жилых площадей прикинем их через полученную пропорцию.
both_known = (df['life_sq'].notna() & df['full_sq'].notna() & df['full_sq'] > 0)
avg_prop = (df.loc[both_known, 'life_sq'] / df.loc[both_known, 'full_sq']).mean()
print(avg_prop)
df.loc[df['life_sq'].isna(), 'life_sq'] = df.loc[df['life_sq'].isna(), 'full_sq'] * avg_prop
print("Количество пропусков после замены:", df['life_sq'].isna().sum())
````

9. Проанализируйте есть ли разница в цене между квартирами на 1 и 9 этажах

````python
selection_1floor = df[(df['floor'] == 1)]
selection_9floor = df[(df['floor'] == 9)]

mean1 = (selection_1floor['price_doc'] / selection_1floor['full_sq']).median()
mean9 = (selection_9floor['price_doc'] / selection_9floor['full_sq']).median()

price_diff = ((mean9 - mean1) / mean1) * 100
if price_diff > 0:
    print(f"9 этаж в среднем дороже на: {price_diff:.2f}%")
elif price_diff < 0:
    print(f"9 этаж в среднем дешевле на: {abs(price_diff):.2f}%")
else:
    print("одинаково")
````

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinski Zaton) с течением времени

````python
import matplotlib.pyplot as plt

selection = df[df['sub_area'] == 'Nagatinskij Zaton'].copy()
selection['timestamp'] = pd.to_datetime(selection['timestamp'])

month_avg = selection.groupby(selection['timestamp'].dt.to_period('M'))['price_doc'].mean()
month_avg.index = month_avg.index.to_timestamp()

# строим график
plt.figure(figsize=(12, 6))
plt.plot(month_avg.index, month_avg.values)
plt.title("Изменение средней цены квартиры в Nagatinski Zaton во времени")
plt.xlabel("Дата")
plt.ylabel("Средняя цена")
plt.show()
````

2. Постройте гистограмму распределения цен на квартиры с общей площадью более 50 кв.метров

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
selection = df[df['full_sq'] < 50].copy()
selection['timestamp'] = pd.to_datetime(selection['timestamp'])
selection = selection[(selection['timestamp'].dt.year >= 2011) & (selection['timestamp'].dt.year <= 2014)]
median_price_monthly = selection.groupby(selection['timestamp'].dt.to_period('M'))['price_doc'].median()
median_price_monthly.index = median_price_monthly.index.to_timestamp()

plt.figure(figsize=(12,6))
plt.plot(median_price_monthly.index, median_price_monthly.values)
plt.title("изменение медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года")
plt.xlabel("Дата")
plt.ylabel("Медианная цена")
plt.grid(True)
plt.show()
````
